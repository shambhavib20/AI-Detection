```python
import pandas as pd
```

```python
data = pd.read_csv('UMS_2.csv')
data.head()
```

```python
print("Number of unique student IDs: ", data['Anon Student Id'].nunique())
```

```python
data.columns
```

```python
filtered_data = data[
    ((data['Source of Truth: Obviously AI-generated'] == 1) & (data['Source of Truth: Authentic, Real Person'] == 0)) |
    ((data['Source of Truth: Obviously AI-generated'] == 0) & (data['Source of Truth: Authentic, Real Person'] == 1)) |
    (data['Source of Truth: Obviously AI-generated'] == 0.5)
]

filtered_data.head()
```

```python
# nans per column
filtered_data.isnull().sum()
```

```python
# IRR 
from sklearn.metrics import cohen_kappa_score
AI_irr = cohen_kappa_score(filtered_data['Rater 1: Obviously AI-generated Response (yes=1; no= 0)'], filtered_data['Rater 2: Obviously AI-generated Response (yes=1; no= 0)'])
human_irr = cohen_kappa_score(filtered_data['Rater 1: Authentic, Real Person Response (yes=1; no=0)'], filtered_data['Rater 2: Authentic, Real Person Response (yes=1; no=0)'])

print("AI IRR: ", AI_irr,
      "\nHuman IRR: ", human_irr)
```

```python
# count 0.5 in dataset
filtered_data['Source of Truth: Obviously AI-generated'].value_counts()
# ai_count = filtered_data['Source of Truth: Obviously AI-generated'].sum()
# human_count = filtered_data['Source of Truth: Authentic, Real Person'].sum()

# print(f"AI-generated responses: {ai_count}")
# print(f"Human-generated responses: {human_count}")
```

```python
from sklearn.model_selection import train_test_split

unique_students = filtered_data['Anon Student Id'].unique()
train_students, test_students = train_test_split(unique_students, test_size=0.2, random_state=42)

train_data = filtered_data[filtered_data['Anon Student Id'].isin(train_students)]
test_data = filtered_data[filtered_data['Anon Student Id'].isin(test_students)]

print(f"Total students: {len(unique_students)}")
print(f"Training students: {len(train_students)}, Training answers: {len(train_data)}")
print(f"Test students: {len(test_students)}, Test answers: {len(test_data)}")

train_data = train_data[['Transaction Id', 'Anon Student Id', 'Input', 'Source of Truth: Obviously AI-generated', ]]
test_data = test_data[['Transaction Id', 'Anon Student Id', 'Input', 'Source of Truth: Obviously AI-generated']]

train_data.to_csv("train_data_5.csv", index=False)
test_data.to_csv("test_data_5.csv", index=False)
```

```python
print('Train set:\n', train_data['Source of Truth: Obviously AI-generated'].value_counts())
print('Test set:\n', test_data['Source of Truth: Obviously AI-generated'].value_counts())
```

```python
import json

def convert_to_jsonl(df, output_file):
    with open(output_file, 'w', encoding='utf-8') as f:
        for _, row in df.iterrows():
            json_obj = {
                "messages": [
                    {"role": "system", "content": """"You are an AI text detection system. Your task is to identify whether a given response was generated by AI, written by a human or if it is uncertain. Here are some characteristics that can help you distinguish:

For AI-generated responses, look for the following features:
- Overly structured response (e.g., bullet points, lengthy or complex sentences with little personal engagement).
- Repetitive phrases or themes
- Very long answers and formal language

For human responses, look for these characteristics:
- Typos, misspellings, or incomplete sentences.
- Casual, conversational tone or slight imperfections in grammar.
- Use of personal phrases like "I would say," "I think," or "I would do."
- Imperfect punctuation or capitalization errors.
- Responses that are more spontaneous or creative
                     
For responses that are uncertain, it would be a mix of the two categories above.

Your task is to classify the following response as either AI-generated or human-generated. Output 1 if the response was obviously generated by AI, and 0 if the response was written by a human. If it is uncertain, output 0.5.

Here is the response to evaluate:
"""},
                    {"role": "user", "content": row['Input']},
                    {"role": "assistant", "content": str(row['Source of Truth: Obviously AI-generated'])}
                ]
            }
            f.write(json.dumps(json_obj) + '\n')

convert_to_jsonl(train_data, "train_5.jsonl")
convert_to_jsonl(test_data, "test_5.jsonl")


# Try with and without the prompt
```

```python
import openai
import os

openai.api_key = os.getenv("OPENAI_API_PLUS")

client = openai.OpenAI(
    api_key=os.getenv("OPENAI_API_PLUS")
)

# print(openai.api_key)
```

```python
training_file = client.files.create(
    file=open("train_5.jsonl", "rb"),
    purpose="fine-tune"
)
file_id = training_file.id
print(file_id)
```

```python
fine_tune = client.fine_tuning.jobs.create(
  training_file= file_id, 
  model="gpt-4o-2024-08-06"
)
print(fine_tune)
```

```python
fine_tune_id = fine_tune.id  
job_status = client.fine_tuning.jobs.retrieve(fine_tune_id)
# print(job_status)
```

```python
fine_tuned_model= job_status.fine_tuned_model
```

```python
prompt = """"You are an AI text detection system. Your task is to identify whether a given response was generated by AI, written by a human or if it is uncertain. Here are some characteristics that can help you distinguish:

For AI-generated responses, look for the following features:
- Overly structured response (e.g., bullet points, lengthy or complex sentences).
- Repetitive phrases or themes
- Very long answers and formal language

For human responses, look for these characteristics:
- Typos, misspellings, or incomplete sentences.
- Casual, conversational tone or slight imperfections in grammar.
- Use of personal phrases like "I would say," "I think," or "I would do."
- Imperfect punctuation or capitalization errors.
- Responses that are more spontaneous or creative
                     
For responses that are uncertain, it would be a mix of the two categories above.

Your task is to:
- Classify the response (1: AI-generated, 0: Human-generated, 0.5: Uncertain).
- Explain the reason for your classification in less than 100 words.
- Ensure the output is in the format: score:reason (1:The response is overly structured and formal with no personal engagement).

Here is the response to evaluate:
"""

def get_prediction(text):
    response = client.chat.completions.create(
        model=fine_tuned_model,
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": f"Classify the following response and explain why in less than 100 words in the format score:reason :\n{text}"}
        ]
    )
    return response.choices[0].message.content 

test_data['Output'] = test_data['Input'].apply(get_prediction)
```

```python
test_data 
```

```python
df = test_data.copy()
df['score'] = df['Output'].str.split(':').str[0]
df['reason'] = df['Output'].str.split(':').str[1]
```

```python
df
```

```python
df['score'] = df['score'].astype(float)
```

```python
print(df['score'].value_counts())
print(df['Source of Truth: Obviously AI-generated'].value_counts())
```

```python
df.to_csv("test_data_with_predictions_5.csv", index=False)
```

```python
df['score'] = df['score'].replace(0.5, 2)
df['Source of Truth: Obviously AI-generated'] = df['Source of Truth: Obviously AI-generated'].replace(0.5, 2)
print(df['score'].value_counts())
print(df['Source of Truth: Obviously AI-generated'].value_counts())
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(df['Source of Truth: Obviously AI-generated'], df['score']))
print(classification_report(df['Source of Truth: Obviously AI-generated'], df['score'], target_names=['Human', 'AI', 'Uncertain']))
```
