```python
import pandas as pd
import requests
```

```python
api_key = ""
url = "https://api.gptzero.me/v2/predict/text"
```

```python
data = pd.read_csv('UMS_2.csv')
data.head()
```

```python
filtered_data = data[
    ((data['Source of Truth: Obviously AI-generated'] == 1) & (data['Source of Truth: Authentic, Real Person'] == 0)) |
    ((data['Source of Truth: Obviously AI-generated'] == 0) & (data['Source of Truth: Authentic, Real Person'] == 1)) |
    (data['Source of Truth: Obviously AI-generated'] == 0.5)
]

filtered_data.head()
```

```python
filtered_data.shape
```

```python
test_data = pd.read_csv('test_data_5.csv')
test_data.shape
```

```python
# random sample of test_data
test_sample = test_data.sample(200, random_state=42)
test_sample['Input'].str.split().str.len().sum()
```

```python
test_sample['Source of Truth: Obviously AI-generated'].value_counts()
```

```python
classifications = []
confidence_scores = []
messages = []

for index, row in filtered_data.iterrows():
    text = row['Input']
    payload = {
        "document": text,
        "multilingual": False
    }
    headers = {
        "x-api-key": api_key,
        "Content-Type": "application/json",
        "Accept": "application/json"
    }
    response = requests.post(url, json=payload, headers=headers)
    output = response.json()

    document = output["documents"][0]  
    predicted_class = document["predicted_class"]  
    confidence_score = document["confidence_score"] 
    result_message = document["result_message"]

    classifications.append(predicted_class)
    confidence_scores.append(confidence_score)
    messages.append(result_message)
```

```python
output
```

```python
df_pred = test_sample.head(164)
```

```python
df_pred['Predicted Class'] = classifications
df_pred['Confidence Score'] = confidence_scores
df_pred['Result Message'] = messages
```

```python
df_pred.to_csv('GPTZero Predictions.csv', index=False)
```

```python
df_pred = pd.read_csv('GPTZero Predictions.csv')
```

```python
df_pred.head()
```

```python
print(df_pred['Predicted Class'].value_counts())
print(df_pred['Source of Truth: Obviously AI-generated'].value_counts())
```

```python
df_pred['Predicted Label'] = df_pred['Predicted Class'].apply(lambda x: 0 if x == 'human' else 1)
df_pred['Label 1 (0.5 = 1)'] = df_pred['Source of Truth: Obviously AI-generated'].apply(lambda x: 1 if x == 0.5 else x) # 0.5 is considered as 1
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(df_pred['Label 1 (0.5 = 1)'], df_pred['Predicted Label']))
print(classification_report(df_pred['Label 1 (0.5 = 1)'], df_pred['Predicted Label']))
```

```python
df_pred['Label 2 (0.5 = 0)'] = df_pred['Source of Truth: Obviously AI-generated'].apply(lambda x: 0 if x == 0.5 else x) # 0.5 = 0

print(confusion_matrix(df_pred['Label 2 (0.5 = 0)'], df_pred['Predicted Label']))
print(classification_report(df_pred['Label 2 (0.5 = 0)'], df_pred['Predicted Label']))
```

```python
df_pred['True Label'] = df_pred['Source of Truth: Obviously AI-generated'].apply(lambda x: 2 if x == 0.5 else x) # 0.5 = 2

print(confusion_matrix(df_pred['True Label'], df_pred['Predicted Label']))
print(classification_report(df_pred['True Label'], df_pred['Predicted Label']))
```
